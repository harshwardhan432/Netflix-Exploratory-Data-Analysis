# pre-processing of data

```python
df.head()
```

```python
df['country']=df['country'].str.split(', ')
df=df.explode('country')
df.head()
```

```python
df['director']=df['director'].str.split(', ')
df=df.explode('director')
df.head()
```

```python
df['listed_in']=df['listed_in'].str.split(', ')
df=df.explode('listed_in')
df.head()
```

```python
df['cast']=df['cast'].str.split(', ')
df=df.explode('cast')
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns=['index'],inplace=True)
```

```python
df.head()
```

```python
for i in df.columns:
    null_rate = df[i].isna().sum()/len(df) * 100
    if null_rate >= 0 :
        print("{} null rate: {}%".format(i,round(null_rate,2)))
```

## Imputing categorical values using value_counts

```python
df['director'] = df['director'].fillna(df.groupby('country')['director'].value_counts().index[0][1])

```

```python
df['cast']=df['cast'].fillna(df.groupby('country')['cast'].value_counts().iloc[0])
```

```python
df['country']=df['country'].fillna(df.groupby('cast')['country'].value_counts().iloc[0])
```

```python
df['date_added']=df['date_added'].fillna(df.groupby('country')['date_added'].value_counts().iloc[0])
```

```python
df['rating']=df['rating'].fillna(df.groupby('country')['rating'].value_counts().iloc[0])
```

```python
df['duration']=df['duration'].fillna(df.groupby('country')['duration'].value_counts().iloc[0])
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df['duration']=df['duration'].str.replace(' min','')
```

```python
df['duration']=df['duration'].str.replace('Seasons','')
```

```python
df['duration']=df['duration'].str.replace('Season','')
```

```python
df['duration']=df['duration'].astype('int')
```

```python
df.info()
```

# Business Insights

### 1.) The top two country of origin is USA and India, so more focus on movies/TV series from this origin
### 2.) International Movies, Dramas, Comedies are the most popular genres of streaming
### 3.) Movies are more watched than TV shows.
### 4.) Release year is thickly concentrated between 2010 to 2020.
### 5.) Most watched TV shows have season length of 1 to 2 seasons and Movies have duration range of 70-200                  mins.
### 6.) Najwa Najjar is the most appeared director globally and David Dhawan is the most appeared director in                     India.
### 7.) Anupam Kher is the most appeared actor in India while Adam Sandler is the most appeared actor globally.
